### Creating and deleting columns

```python
movie = pd.read_csv('data/movie.csv')
movie['has_seen'] = 0
movie['any'] = round(movie.movie_facebook_likes/30,2)
movie.head()
```

```python
# defining function to check price 
def fun_1(num): 
  
    if num<=10: 
        return "10  or Less"
  
    elif num > 10 and num<=20: 
        return "10 to 20 "
    elif num > 20 and num<=30: 
        return "20 to 30 "
    elif num > 30 and num<=40: 
        return "30 to 40 "
    elif num > 40 and num<=50: 
        return "40 to 50 "
    elif num > 50 and num<=60: 
        return "50 to 60 "
  
    elif num > 60 : 
        return "Greater Than 60 "
```

```python
#movie['tanvi'] = movie.director_facebook_likes/10
movie['like_group'] = movie['director_facebook_likes'].apply(fun_1) 
movie.like_group.value_counts()
```

```python
movie.columns
```

```python
movie['actor_director_facebook_likes'] = (movie['actor_1_facebook_likes'] + 
                                              movie['actor_2_facebook_likes'] + 
                                              movie['actor_3_facebook_likes'] + 
                                              movie['director_facebook_likes'])
```

```python
movie['actor_director_facebook_likes'].isnull().sum()
```

```python
movie = movie.drop(['actor_director_facebook_likes','any'], axis='columns')
movie.columns
```

```python
movie['actor_total_facebook_likes'] = (movie['actor_1_facebook_likes'] + 
                                       movie['actor_2_facebook_likes'] + 
                                       movie['actor_3_facebook_likes'])

movie['actor_total_facebook_likes'] = movie['actor_total_facebook_likes'].fillna(0)
```

```python
movie['is_cast_likes_more'] = movie['cast_total_facebook_likes'] >= \
                                  movie['actor_total_facebook_likes']
    
movie['is_cast_likes_more'].all()
```
